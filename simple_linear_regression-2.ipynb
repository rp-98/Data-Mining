```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv('cars.csv')
X=df[['speed']]
Y=df['dist']
#df.describe()
#df.info
#df2
#df.dtypes
#df.columns
#df = df.reindex(np.random.permutation(df.index))
#df.head()
```

```python
plt.plot(X,Y,'r*')
plt.title('speed Vs Distance')
plt.xlabel('X-speed')
plt.ylabel('Y-distance')
sb.set_style("dark")
plt.show()
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X,Y)
estimates=(model.intercept_,model.coef_)
print(estimates)
pred=model.predict(X)
error=pred-Y
plt.plot(X,pred,'r')
plt.plot(X,Y,'*')
plt.title('speed Vs Dist regression')
#plt.title('predicted dist V/s actual dist')
plt.xlabel('X-speed')
plt.ylabel('Y-distance')
sb.set_style("dark")
plt.grid()
#=['predited', 'actual']
#plt.legend()
```

```python
plt.plot(X,error,'r*')
plt.title('scatter plot of residuals ')
```

```python
import statsmodels.api as sm
X = sm.add_constant(X)
regr = sm.OLS(Y,X)
model = regr.fit()
```

```python
print(model.summary())
```

```python
#QQ plot
fig=sm.qqplot(error)
plt.show()
```

```python
#histogram of residuals
plt.hist(error)
plt.title('distribution of residuls')
plt.show()
```

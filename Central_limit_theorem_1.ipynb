```python
import numpy as np
import matplotlib.pyplot as plt
rnd= np.random.rand(100000)*10 # population of random number b/w 0 and 10
mean_p=np.mean(rnd)  # population mean
sd_p=np.std(rnd)     #population std
print('population_mean =  ', mean_p)
print('population_std =   ',sd_p)
```

```python
n=900  # sample size  n=25,100,225, 400 ,625,900
```

```python
import random
means=[]
sample_sd=[]
for i in range(1000):
    sample=random.sample(list(rnd),25)   # random sampling
    means.append(np.mean(sample)) # sample mean
    sample_sd.append(np.std(sample)) # sample std
#means of collected sample mean
mean_m=np.mean(means)
#std of collected std sample
sd_m= np.std(means)
sd_smp= sd_p/np.sqrt(n)
print('mean_m =',mean_m)
print('sd_m =', sd_m)
print('sd_smp =', sd_smp)
```

```python
print(sample_sd)
```

```python
#histograms of means and  sample_sd
plt.hist(means)
plt.title('distribution of sample means /'+'n='+str(n))
plt.show()
plt.hist(sample_sd)
plt.title('distribution of sample std/'+'n='+str(n))
plt.show()
```
